The 'Brief Sample Overview' gives a generic oversight on the sample population as a whole. However, in oder to to derive valuable insight, the sample must be explored further. To achieve this, the sample will be split into groups based on their age and gender for the rest of section 1.

```python
# This cell creates two separate dataframe's. One for males and one for females 

# Create two empty lists for data to be stored into
male_data_list = []
female_data_list = []

# Using same for loop as before to go through each row of the dataset individually
for index, row in gym_members_data.iterrows():
    # If statement to check gender and separate based on result
    if row['Gender'] == 'Male':
        male_data_list.append(row)  # Add male rows data to male_data_list
    elif row['Gender'] == 'Female':
        female_data_list.append(row)  # Add female rows data to female_data_list

# Next the data is converted from the lists back to dataframes
male_data = pd.DataFrame(male_data_list, columns=gym_members_data.columns) # This makes sure the column names are consistant
female_data = pd.DataFrame(female_data_list, columns=gym_members_data.columns)

# Print the new dataframes
print("Male Data:")
print(male_data)

print("\nFemale Data:")
print(female_data)
```

```python
# This cell breaks down the new male and female dataframes into age groups to understand the partcipant group further

# As these age brackets will be called upon multiple times in this section, they have been turned into a function
def age_brackets(age): # Setting parameter as 'age' as that is the metric being looked at
    if 18 <= age <= 24: # Using '<=' rather than just '<' means that if somebody is 18 or 24 they still meet the condition 
        return '18-24' 
        # 'return' passes the age bracket that a gym member belongs to outside the funtion as a string - this will be useful later
    elif 25 <= age <= 34:
        return '25-34'
    elif 35 <= age <= 44:
        return '35-44'
    elif 45 <= age <= 54:
        return '45-54'
    else:           # 'Else' gathers up all the partcipants that do not meet the previous 'if' or 'elif' conditions
        return '55+'


def calculate_age_distribution(data): # Setting paramter as 'data' as this is where the required dataframe will go when calling the function
    # Setting age distribution counters to 0 to be added to as partcipants meet conditions set in 'age_bracket' function
    age_ranges_counter = {'18-24': 0, '25-34': 0, '35-44': 0, '45-54': 0, '55+': 0}
    
    for index, row in data.iterrows():
        members_age_brackets = age_brackets(row['Age']) # Using variable to recall the 'age_brackets' function with defined age ranges
        age_ranges_counter[members_age_brackets] += 1 
        # As the function returns values based on the member's age, the corresponding age bracket in the 'age_ranges_counter' dictionary
        # has 1 added to its current value
            
    return age_ranges_counter # 'return' sends the values collected by the counter outside the function so they can be called upon

# Create variables to store the calculated values of the recalled functions
male_age_ranges = calculate_age_distribution(male_data)
female_age_ranges = calculate_age_distribution(female_data)

# Print results variables above
print("Age Distribution for Males:", male_age_ranges, "\n")
print("Age Distribution for Females:", female_age_ranges)
```

```python
# This cell aims to find out the most common workout type each gender undergos

def workout_preferences(dataframe, gender):
    
    # Here is an empty dictionary to store the keys (workout types) and values (how many partcipant's choose that workout types)
    workout_count = {}

    # Creating another dictionary with keys as the selected age brackets but nested dictionaries within to store the wokout types and counts
    age_based_workout_count = {'18-24': {}, '25-34': {}, '35-44': {}, '45-54': {}, '55+': {}}

    for index, row in dataframe.iterrows():
        workout_type = row['Workout_Type']

        # Counting the overall workout types
        if workout_type not in workout_count: # 'not in' checks that the workout type isn't already in the dictionary and adds it if not
            workout_count[workout_type] = 1
        # Adding workout type to dictionary with a value of 1 as if it wasn't already in there, then this is the first person to prefer it
        else:
            workout_count[workout_type] += 1
        
        # Determine age bracket of the participant
        age_bracket = age_brackets(row['Age'])
        
         # This 'if' statement repeats the same process as the above but this time sends values to the 'age_based_workout_count' dictionary
        if workout_type not in age_based_workout_count[age_bracket]:
            # Recall function from last cell to determine which nested dictionary the workout type is being added to
            age_based_workout_count[age_bracket][workout_type] = 1
        else:
            age_based_workout_count[age_bracket][workout_type] += 1

    # Determines the most common workout overall
    # Source : (Python)
    most_common_workout = max(workout_count, key=workout_count.get)
    # max() finds the workout type with the highest value count from 'workout_count' dictionary - as specified by the '.get' method
    print(f"Most Common Workout Overall for {gender}s:\n  {most_common_workout}\n")

    # Determine the most common workout for each age bracket
    print(f"Most Common Workout for {gender}s by Age Bracket:")
    for age_bracket, workout_data in age_based_workout_count.items():
        # Must use a loop to go through all the items in 'age_based_workout_count' dictionary
        most_common_workout = max(workout_data, key=workout_data.get)
        print(f"  {age_bracket}: {most_common_workout}\n")
     

# Apply the function to different genders' dataframes
workout_preferences(male_data, "Male")
workout_preferences(female_data, "Female")

# Following the results from the data this section was created to understand the feamle 45-54 bracket better:

# Create empty dictionary to store keys and values
female_45_54_workout_count = {}

for index, row in female_data.iterrows():
    # Only looking at women aged 45-54
    if 45 <= row['Age'] <= 54:
        workout_type = row['Workout_Type']

        if workout_type not in female_45_54_workout_count:
            female_45_54_workout_count[workout_type] = 1
        else:
            female_45_54_workout_count[workout_type] += 1

print("Workout counts for females aged 45-54:")
for workout_type, count in female_45_54_workout_count.items():
    print(f"{workout_type}: {count}")
```

```python
def average_workout_frequency_by_age(dataframe, gender):

    workout_freq = {'total_workout_frequency': 0,'count': 0}
    
    # Initialises a dictionary to track workout data by age bracket based of dataframe selected
    age_bracket_workout_freq_data = {
    '18-24': {'total_workout_frequency': 0, 'count': 0},
    '25-34': {'total_workout_frequency': 0, 'count': 0},
    '35-44': {'total_workout_frequency': 0, 'count': 0},
    '45-54': {'total_workout_frequency': 0, 'count': 0},
    '55+': {'total_workout_frequency': 0, 'count': 0}
}
    
    for index, row in dataframe.iterrows():
        
        workout_freq['total_workout_frequency'] += row['Workout_Frequency (days/week)']
        # Value associated with 'total_workout_frequency' key increases by 'Workout_Frequency' figure
        workout_freq['count'] += 1
        
        age_bracket = age_brackets(row['Age']) # Determine the age bracket by recalling predefined 'age_brackets' function
        
        age_bracket_workout_freq_data[age_bracket]['total_workout_frequency'] += row['Workout_Frequency (days/week)']
        # Same calculations s above but this time they are associated with the age bracket the member belongs to
        age_bracket_workout_freq_data[age_bracket]['count'] += 1

    # Prints the results for each gender as a whole
    print(f"Average Workout Frequency for {gender}s as a Collective:")
    overall_average = round(workout_freq['total_workout_frequency'] / workout_freq['count'], 2)
    print(f"  {overall_average} days/week \n")
    
    # Prints the average workout frequency for each age bracket
    print(f"Average Workout Frequency for {gender}s by Age Bracket:")
    for age_bracket, data in age_bracket_workout_freq_data.items():
        average_frequency = round(data['total_workout_frequency'] / data['count'], 2)
        print(f"  {age_bracket}: {average_frequency} days/week \n" )


# Apply the function to male and female datasets
average_workout_frequency_by_age(male_data, "Male")
average_workout_frequency_by_age(female_data, "Female")
```

```python
# This cell looks into the average session duration of male and female gym members

def average_session_duration_by_age(dataframe, gender):

    # Follows the same pattern as the dictionaries from the previous cell but this time for session durations
    session_duration = {'total_session_duration': 0, 'count': 0}
    
    # Track session duration data by age bracket 
    age_bracket_session_duration_data = {
        '18-24': {'total_session_duration': 0, 'count': 0},
        '25-34': {'total_session_duration': 0, 'count': 0},
        '35-44': {'total_session_duration': 0, 'count': 0},
        '45-54': {'total_session_duration': 0, 'count': 0},
        '55+': {'total_session_duration': 0, 'count': 0}
    }
    
    for index, row in dataframe.iterrows():
        
        session_duration['total_session_duration'] += row['Session_Duration (hours)']
        # Increase total session duration for all gym members
        session_duration['count'] += 1
        
        age_bracket = age_brackets(row['Age']) # Determine the age bracket by recalling predefined 'age_brackets' function
        
        age_bracket_session_duration_data[age_bracket]['total_session_duration'] += row['Session_Duration (hours)']
        # Increment the total session duration and count for the relevant age bracket
        age_bracket_session_duration_data[age_bracket]['count'] += 1

    # Print the results for each gender as a whole
    print(f"Average Session Duration for {gender}s as a Collective:")
    overall_average = round(session_duration['total_session_duration'] / session_duration['count'], 2)
    print(f"  {overall_average} hours/session \n")
    
    # Print the average session duration for each age bracket
    print(f"Average Session Duration for {gender}s by Age Bracket:")
    for age_bracket, data in age_bracket_session_duration_data.items():
        average_duration = round(data['total_session_duration'] / data['count'], 2)
        print(f"  {age_bracket}: {average_duration} hours/session \n" )


# Apply the function to male and female datasets
average_session_duration_by_age(male_data, "Male")
average_session_duration_by_age(female_data, "Female")
```

```python
# This cell aggregates the BMI average and outliers of both male and females

def bmi_statistics(dataframe, gender): 
    # Once again set counters to 0
    total_bmi = 0         # Variable to calculate average
    outliers = 0          # Stores the number of gym members with BMI < 18.5 or > 30
    underweight = 0       # Stores the number of underweight males (under 18.5 BMI)
    obese = 0             # Stores the number of overweight males (over 30 BMI)
    count = 0             

    for index, row in dataframe.iterrows():
        total_bmi += row['BMI']  
        if row['BMI'] < 18.5:    # If the gym member has a BMI below 18.5 then both the below variables value increases by 1
            underweight += 1
            outliers += 1        
        elif row['BMI'] > 30:    # If the male gym member has a BMI above 30 then both the below variables value increases by 1
            obese += 1
            outliers += 1        
        count += 1              

    # Calculate the average BMI, rounded to 2 decimal places and print all results
    average_bmi = round(total_bmi / count, 2)
    print(f"\nThe average BMI for {gender}s is:", average_bmi)
    print(f"Total number of BMI Outliers for {gender}s:", outliers)
    print(f"Number of Underweight {gender}s (BMI < 18.5):", underweight)
    print(f"Number of Obese {gender}s (BMI > 30): {obese} \n")

# Apply the function to male and female datasets
bmi_statistics(male_data, "Male")
bmi_statistics(female_data, "Female")


# After seeing initial results this section was created to get further insight
# Created a dictionary with the diffrent age groups as keys and associated values starting at 0
underweight_female_age_brackets = {
    "18-24": 0,
    "25-34": 0,
    "35-44": 0,
    "45-54": 0,
    "55+": 0
}

#For loop to go through the female dataframe and add values associated with relevant keys if the condition is met
for index, row in female_data.iterrows():
    if row['BMI'] < 18.5:  # Check if the BMI is underweight
        underweight_females = age_brackets(row['Age'])  # Determine the age bracket  by recalling predetermined function
        underweight_female_age_brackets[underweight_females] += 1

# Returns the summed up values with their associated age bracket
print("The number of underweight females in each age bracket are as follows:")
for underweight_females, underweight_count in underweight_female_age_brackets.items(): # .items() works in the same way as .iterrows but for dictionaries
    print(f"{underweight_females}: {underweight_count}")

```

```python
# This looks at the distribution of obese males accordng to BMI and males who are actually obese

# Create the empty dictionaries to store the values in 
bmi_obese_male_age_brackets = {
    "18-24": 0,
    "25-34": 0,
    "35-44": 0,
    "45-54": 0,
    "55+": 0
}

actually_obese_male_age_brackets = {
    "18-24": 0,
    "25-34": 0,
    "35-44": 0,
    "45-54": 0,
    "55+": 0
}

# Loop through the male dataframe
for index, row in male_data.iterrows():
    age_bracket = age_brackets(row['Age'])  # Determine the age bracket using the predefined 'age_brackets' function
    
    # If BMI is greater than 30, increase the counter for the age bracket
    if row['BMI'] > 30:
        bmi_obese_male_age_brackets[age_bracket] += 1
        
        # If BMI > 30 and Fat Percentage > 25, increase counter for the age bracket
        if row['Fat_Percentage'] > 25:
            actually_obese_male_age_brackets[age_bracket] += 1

# Prints results
print("The number of obese males (BMI > 30) in each age bracket are as follows:")
for age_bracket, count in bmi_obese_male_age_brackets.items():
    print(f"{age_bracket}: {count} \n")

print("\nThe number of obese males with BMI > 30 and Fat Percentage > 25 in each age bracket are as follows:")
for age_bracket, count in actually_obese_male_age_brackets.items():
    print(f"{age_bracket}: {count} \n")
```

```python
#This cell introduces NumPy and utilises its statistical functions to gain inisght on "workout_type" vs "Calories_per_Minute"

import numpy as np

# Makes a function that returns key numpy statistics for a specific column in the selected dataframe to be analysed
def numpy_statistics(dataframe, column):
    return {
        'Mean': np.mean(dataframe[column]),
        'Median': np.median(dataframe[column]),
        'Standard Deviation': np.std(dataframe[column]),
        'Min': np.min(dataframe[column]),
        'Max': np.max(dataframe[column])
    }

gym_members_data = pd.DataFrame(gym_members_data)

# Group data by Workout_Type
grouped_workout_types = gym_members_data.groupby('Workout_Type')

# Calculate statistics for each group and store them in a dictionary
results = {}
for workout_type, group in grouped_workout_types:
    # Apply numpy_statistics to the "Calories_per_Minute" column for the current group
    results[workout_type] = numpy_statistics(group, 'Calories_per_Minute')

print(f"The Following Represnts Calories Burned per Minute:\n")
# Print the results in a formatted manner
for workout_type, stats in results.items():
    print(f"Workout Type: {workout_type}")
    for stat, value in stats.items():
        print(f"{stat}: {round(value,2)}")
    print('\n')  # Add separator line between workout types
```

```python
# This cell runs a one way ANOVA test from Scipy.stats to see if there is a relationship between experience level and calories burned

# Source: (SciPy)
# Performs one-way ANOVA to test if average Calories_per_Minute differs significantly when compared to Experience_Level.
from scipy.stats import f_oneway

# The following block of code explanation:
# Using a FOR loop and .unique() code itterates through the dataset identifying sll unique experience levels (ie 1 2 3)
# For any row where the experience level = a certain level (for example 1), the Calories_per_Minute is fetched and stored in an array
# This repeated for other unique values
# * fetches all unique value arrays for the f_oneway to calculate the F-Statistic and P-Value which are store in the varaibles beofre the '='
# Then an f string is used for ease of printing a mixed of text string and variables with stored values

def ANOVA_test(column):
    f_stat, p_value = f_oneway(   
        *[gym_members_data[gym_members_data['Experience_Level'] == level][column] 
            for level in gym_members_data['Experience_Level'].unique()]) 
    print(f"F-Statistic {column}: {f_stat}, P-Value: {p_value}")

ANOVA_test('Calories_per_Minute')
ANOVA_test('Total_Calories_Burned_per_Week')
```

The one way ANOVA function used above is to test how significant of an effect experience level has on the variables above. The test results return an F-Statistic and P-Value which are the.

The F-Statistic indicates the relative size differences between groups in contrast to the variations within groups. A higher F-Statistic suggests that the differences between different group figures are significant (there are noticeable differences between groups) which in this case would mean experience level might have a meaningful effect on the calories burned.

The P-Value helps determine whether the observed differences are statistically significant which refers to the likelihood the results observed in the data are not due to chance, but instead reflect a real relationship. In terms of significance, 0.05 is commonly regarded as the highest number a P-Value can be for there to be a reltionship between variables.

The initial test between experience level and total calories burned per minute retruns an F-Statistic of 0.186 (very small) and a P-Value of 0.83 (much larger than 0.05) therefore indidicating that there is no reltionship between how experienced someone is and how many calories they burn per minute. However the following test rveals rhe complete inverse, with an F-Statistic of 1630.1 and P-Value of 6.33 x 10 to the power of -311, there is an undeniable relationship between how experienced someone is and how many calories they burn in a week.

Due to the disparity in results, it is clear that there are more variables in play when looking at why gym members with more experience burn more calories in a week than those of less experience as there is no noticeable difference in how many claories per minute they tend to burn. Examples of such variables are likely the more experienced members tend to work out longer or more frequently than those of less experience. 

```python
# This cell is to see if those who are experinced members tend to workout for longer or more frequently

ANOVA_test('Session_Duration (hours)')
ANOVA_test('Workout_Frequency (days/week)')
```

The Pearson correlation test is a statistical method used to measure the strength and direction of the relationship between continuous variables. The result of the test ranges from -1 to +1; a value of +1 indicating a perfect positive linear relationship (both variables increase together) and a value of -1 indicating a perfect negative linear relationship  (one variable increases while the other decreases). Lastly, a value of 0 means there is no linear relationship between the variables. 
Like with the ANOVA one way test, a p-value is also calculated to determine if the observed correlation is statistically significant, with values less than 0.05 once more indicating a correlation is unlikely to have occurred by chance.

When faced with missing values, multiple options present themselves. However for this notebook only two are feasible: removing all the rows with missing values or replacing the missing values with mean, median or mode. If all the rows with data missing were removed, data on more than just just the missing cells would be lost and with such a large proportion of the sample missing data, this option would lead to a significant loss of valuable information. Therefore, the more practical solution for this analysis is to replace the missing values with the mean or median or mode of the respective columns. 

If the values in the columns with missing values had extreme outliers, the median would likely be the best method to replace missing values as it is more robust against outliers as shown in the previous results in the section. Furthermore, mode would be the preferred choice if there was a repeated data point that would represent the colmn as a whole. However as the data in these columns is continuous and lacks outliers, mean will be the preferred replacemnet method for the next cell. 

```python
# This cell replaces all the missing values with the mean of the repective columns

# Makes a function that replaces missing values with the mean for the specified column
def mean_of_missing_data (column):
    gym_members_data[column] = gym_members_data[column].fillna(
        gym_members_data[column].mean() # Using mean for reasons justified in markdown above
)

# Recalls function to use on these columns
mean_of_missing_data('Fat_Percentage')
mean_of_missing_data('Water_Intake (liters)')

# Displays the specific rows that used to have missing values
updated_rows = gym_members_data.loc[[4, 13, 33, 51, 59], ['Fat_Percentage', 'Water_Intake (liters)']]
print(updated_rows)


# Sums up the amount of missing values left in each column to double check the missing values are replaced
print("\nCells with Missing Values per Column:")
print(gym_members_data[['Fat_Percentage', 'Water_Intake (liters)']].isnull().sum())
```

Notably, both yoga and HIIT training have a denser population of gym members in the 10-15% fat percentage range when compared to strength training and cardio. This might be expected due to the high-intensity nature of HIIT training (which is highly fat-burning); however, yoga came as a surprise. This may have been an oversight, considering yoga is centralised around manoeuvring body weight. The majority of gym members seem to be within the 25-30% fat percentage range, with the highest density preferring cardio workouts.

Graphs 2 and 3 plot gym members' fat percentage statistics against total calories burned in a week and water intake, respectively. Based on the distribution of the scatter points and the regression line, Graph 2 clearly displays a negative linear relationship between fat percentage and calories burnedthe more calories burned in a week, the lower the fat percentage. Interestingly, the 4,000 calories burned per week mark seems to be decisive when it comes to fat percentage, with the majority of those who burn under that amount having a body fat percentage of 20% or higher, whereas the majority who burn over that amount tend to be under 20% body fat. No one with a body fat percentage of around 35% exceeds the 4,000-calorie mark. There are exceptions, however, such as one individual who burns over 6,000 calories per week but still has 25% body fat. This is not an unhealthy fat percentage for either gender but indicates that other lifestyle choices may be contributing to this individuals higher fat percentage than would be expected.

Graph 3 also displays a negative linear relationship, but between fat percentage and water intakethe more water consumed in a day, the lower the fat percentage. Water consumption is minimal in calories and improves body functions such as blood circulation and metabolic rates, so this is not surprising. Additionally, those who tend to exercise more, and therefore burn more fat, are likely to be the ones drinking the most water to rehydrate during or after exercise. What is noticeable from the graph, though, is that although there are two outliers at around 2.75 litres a day, the majority of gym members with a body fat percentage below 15% drink exactly 3.5 litres a day. Perhaps there is a sports science reason for this, or it could simply be because they exercise the most and need to replenish more of their body fluids.

Graphs 4 and 5 use Kernel Density Estimate (KDE) plots to show the density of gym members' fat percentage in comparison to their resting BPM and the effort exerted during a gym session, as measured by the average increase in heart rate. There are multiple factors that can affect one's resting heart rate, such as stress, age, and medication, but what is generally considered average is around 60-70 BPM. This is reflected in Graph 4, where the deepest blue is found in the range where the average heart rate meets the average fat percentage. It was expected that a larger proportion of those with 15% or less body fat would have a lower resting heart rate; however, there is actually a higher density of people within the 65-75 BPM range than anywhere else in that fat percentage bracket.

Graph 5 takes a slightly different shape, showing its deepest blue between 25-30% body fat, where the average heart rate increase is between 100-150%. As someone does more exercise, their cardiovascular system improves, leading to a lower increase in heart rate when performing exercise. This is shown in the data, as no one with a body fat percentage below roughly 17.5% reaches a 200% increase, whereas there is a segment of the sample from 22.5% body fat onwards that do. This same logic, though, also explains why this may not be the best indicator of effort. For someone with 35%+ body fat, raising their heart rate by 100% or even 150% will take far less exertion than for someone who is more athletic.

Lastly, Graph 6 uses a box plot to show the distribution of body fat percentages across various experience levels of gym members. Immediately, it is clear that the most experienced members in the gym have the lowest body fat percentages, with a median just under 15% and an upper whisker that only just reaches the lower quartile of the intermediate gym members. The differences between intermediate members and beginners are less than might be expected, with almost identical upper and lower quartile ranges. However, both groups mostly sit within the 25-30% body fat percentage range, yet their upper and lower whiskers are also very similar, with the only real difference being that the intermediate members have a slightly lower mean of roughly 27%. What one might consider a "beginner" may differ from another person's interpretation of "intermediate," which could have impacted the results here. Nevertheless, the key insightfrom this graph is that the more experienced members tend to have the lowest body fat.

The biggest challenge in its creation came in the form of the presentation and functionality of the graph window. After implementing dropdown menus in the aggregation window, it seemed intuitive to use the same approach in the graph window to ensure a simple and consistent user experience. Much like the other pages, activating the functions on both pages also requires the use of a button. This is where the issue arose. After pressing the button, the graph would consume so much of the window that the button would disappear entirely. This meant that if the user wanted to change the metrics used to create the graph, they could not do so, as the button was covered.

To resolve this, the dimensions of the graph's frame were adjusted, and the padding between elements was tweaked. The next issue was ensuring that a new graph would appear once the selected metrics changed. This was resolved more quickly by using a for loop and the .winfo_children() method, which dynamically clears the previous graph from the frame before rendering the new one.
