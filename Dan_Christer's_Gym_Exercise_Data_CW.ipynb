The 'Brief Sample Overview' gives a generic oversight on the sample population as a whole. However, in oder to to derive valuable insight, the sample must be explored further. To achieve this, the sample will be split into groups based on their age and gender for the rest of section 1.

```python
# This cell creates two separate dataframe's. One for males and one for females 

# Create two empty lists for data to be stored into
male_data_list = []
female_data_list = []

# Using same for loop as before to go through each row of the dataset individually
for index, row in gym_members_data.iterrows():
    # If statement to check gender and separate based on result
    if row['Gender'] == 'Male':
        male_data_list.append(row)  # Add male rows data to male_data_list
    elif row['Gender'] == 'Female':
        female_data_list.append(row)  # Add female rows data to female_data_list

# Next the data is converted from the lists back to dataframes
male_data = pd.DataFrame(male_data_list, columns=gym_members_data.columns) # This makes sure the column names are consistant
female_data = pd.DataFrame(female_data_list, columns=gym_members_data.columns)

# Print the new dataframes
print("Male Data:")
print(male_data)

print("\nFemale Data:")
print(female_data)
```

```python
# This cell breaks down the new male and female dataframes into age groups to understand the partcipant group further

# As these age brackets will be called upon multiple times in this section, they have been turned into a function
def age_brackets(age): # Setting parameter as 'age' as that is the metric being looked at
    if 18 <= age <= 24: # Using '<=' rather than just '<' means that if somebody is 18 or 24 they still meet the condition 
        return '18-24' 
        # 'return' passes the age bracket that a gym member belongs to outside the funtion as a string - this will be useful later
    elif 25 <= age <= 34:
        return '25-34'
    elif 35 <= age <= 44:
        return '35-44'
    elif 45 <= age <= 54:
        return '45-54'
    else:           # 'Else' gathers up all the partcipants that do not meet the previous 'if' or 'elif' conditions
        return '55+'


def calculate_age_distribution(data): # Setting paramter as 'data' as this is where the required dataframe will go when calling the function
    # Setting age distribution counters to 0 to be added to as partcipants meet conditions set in 'age_bracket' function
    age_ranges_counter = {'18-24': 0, '25-34': 0, '35-44': 0, '45-54': 0, '55+': 0}
    
    for index, row in data.iterrows():
        members_age_brackets = age_brackets(row['Age']) # Using variable to recall the 'age_brackets' function with defined age ranges
        age_ranges_counter[members_age_brackets] += 1 
        # As the function returns values based on the member's age, the corresponding age bracket in the 'age_ranges_counter' dictionary
        # has 1 added to its current value
            
    return age_ranges_counter # 'return' sends the values collected by the counter outside the function so they can be called upon

# Create variables to store the calculated values of the recalled functions
male_age_ranges = calculate_age_distribution(male_data)
female_age_ranges = calculate_age_distribution(female_data)

# Print results variables above
print("Age Distribution for Males:", male_age_ranges, "\n")
print("Age Distribution for Females:", female_age_ranges)
```

```python
# This cell aims to find out the most common workout type each gender undergos

def workout_preferences(dataframe, gender):
    
    # Here is an empty dictionary to store the keys (workout types) and values (how many partcipant's choose that workout types)
    workout_count = {}

    # Creating another dictionary with keys as the selected age brackets but nested dictionaries within to store the wokout types and counts
    age_based_workout_count = {'18-24': {}, '25-34': {}, '35-44': {}, '45-54': {}, '55+': {}}

    for index, row in dataframe.iterrows():
        workout_type = row['Workout_Type']

        # Counting the overall workout types
        if workout_type not in workout_count: # 'not in' checks that the workout type isn't already in the dictionary and adds it if not
            workout_count[workout_type] = 1
        # Adding workout type to dictionary with a value of 1 as if it wasn't already in there, then this is the first person to prefer it
        else:
            workout_count[workout_type] += 1
        
        # Determine age bracket of the participant
        age_bracket = age_brackets(row['Age'])
        
         # This 'if' statement repeats the same process as the above but this time sends values to the 'age_based_workout_count' dictionary
        if workout_type not in age_based_workout_count[age_bracket]:
            # Recall function from last cell to determine which nested dictionary the workout type is being added to
            age_based_workout_count[age_bracket][workout_type] = 1
        else:
            age_based_workout_count[age_bracket][workout_type] += 1

    # Determines the most common workout overall
    most_common_workout = max(workout_count, key=workout_count.get)
    # max() finds the workout type with the highest value count from 'workout_count' dictionary - as specified by the '.get' method
    print(f"Most Common Workout Overall for {gender}s:\n  {most_common_workout}\n")

    # Determine the most common workout for each age bracket
    print(f"Most Common Workout for {gender}s by Age Bracket:")
    for age_bracket, workout_data in age_based_workout_count.items():
        # Must use a loop to go through all the items in 'age_based_workout_count' dictionary
        most_common_workout = max(workout_data, key=workout_data.get)
        print(f"  {age_bracket}: {most_common_workout}\n")
     

# Apply the function to different genders' dataframes
workout_preferences(male_data, "Male")
workout_preferences(female_data, "Female")

# Following the results from the data this section was created to understand the feamle 45-54 bracket better:

# Create empty dictionary to store keys and values
female_45_54_workout_count = {}

for index, row in female_data.iterrows():
    # Only looking at women aged 45-54
    if 45 <= row['Age'] <= 54:
        workout_type = row['Workout_Type']

        if workout_type not in female_45_54_workout_count:
            female_45_54_workout_count[workout_type] = 1
        else:
            female_45_54_workout_count[workout_type] += 1

print("Workout counts for females aged 45-54:")
for workout_type, count in female_45_54_workout_count.items():
    print(f"{workout_type}: {count}")
```

```python
def average_workout_frequency_by_age(dataframe, gender):

    workout_freq = {'total_workout_frequency': 0,'count': 0}
    
    # Initialises a dictionary to track workout data by age bracket based of dataframe selected
    age_bracket_workout_freq_data = {
    '18-24': {'total_workout_frequency': 0, 'count': 0},
    '25-34': {'total_workout_frequency': 0, 'count': 0},
    '35-44': {'total_workout_frequency': 0, 'count': 0},
    '45-54': {'total_workout_frequency': 0, 'count': 0},
    '55+': {'total_workout_frequency': 0, 'count': 0}
}
    
    for index, row in dataframe.iterrows():
        
        workout_freq['total_workout_frequency'] += row['Workout_Frequency (days/week)']
        # Value associated with 'total_workout_frequency' key increases by 'Workout_Frequency' figure
        workout_freq['count'] += 1
        
        age_bracket = age_brackets(row['Age']) # Determine the age bracket by recalling predefined 'age_brackets' function
        
        age_bracket_workout_freq_data[age_bracket]['total_workout_frequency'] += row['Workout_Frequency (days/week)']
        # Same calculations s above but this time they are associated with the age bracket the member belongs to
        age_bracket_workout_freq_data[age_bracket]['count'] += 1

    # Prints the results for each gender as a whole
    print(f"Average Workout Frequency for {gender}s as a Collective:")
    overall_average = round(workout_freq['total_workout_frequency'] / workout_freq['count'], 2)
    print(f"  {overall_average} days/week \n")
    
    # Prints the average workout frequency for each age bracket
    print(f"Average Workout Frequency for {gender}s by Age Bracket:")
    for age_bracket, data in age_bracket_workout_freq_data.items():
        average_frequency = round(data['total_workout_frequency'] / data['count'], 2)
        print(f"  {age_bracket}: {average_frequency} days/week \n" )


# Apply the function to male and female datasets
average_workout_frequency_by_age(male_data, "Male")
average_workout_frequency_by_age(female_data, "Female")
```

```python
# This cell looks into the average session duration of male and female gym members

def average_session_duration_by_age(dataframe, gender):

    # Follows the same pattern as the dictionaries from the previous cell but this time for session durations
    session_duration = {'total_session_duration': 0, 'count': 0}
    
    # Track session duration data by age bracket 
    age_bracket_session_duration_data = {
        '18-24': {'total_session_duration': 0, 'count': 0},
        '25-34': {'total_session_duration': 0, 'count': 0},
        '35-44': {'total_session_duration': 0, 'count': 0},
        '45-54': {'total_session_duration': 0, 'count': 0},
        '55+': {'total_session_duration': 0, 'count': 0}
    }
    
    for index, row in dataframe.iterrows():
        
        session_duration['total_session_duration'] += row['Session_Duration (hours)']
        # Increase total session duration for all gym members
        session_duration['count'] += 1
        
        age_bracket = age_brackets(row['Age']) # Determine the age bracket by recalling predefined 'age_brackets' function
        
        age_bracket_session_duration_data[age_bracket]['total_session_duration'] += row['Session_Duration (hours)']
        # Increment the total session duration and count for the relevant age bracket
        age_bracket_session_duration_data[age_bracket]['count'] += 1

    # Print the results for each gender as a whole
    print(f"Average Session Duration for {gender}s as a Collective:")
    overall_average = round(session_duration['total_session_duration'] / session_duration['count'], 2)
    print(f"  {overall_average} hours/session \n")
    
    # Print the average session duration for each age bracket
    print(f"Average Session Duration for {gender}s by Age Bracket:")
    for age_bracket, data in age_bracket_session_duration_data.items():
        average_duration = round(data['total_session_duration'] / data['count'], 2)
        print(f"  {age_bracket}: {average_duration} hours/session \n" )


# Apply the function to male and female datasets
average_session_duration_by_age(male_data, "Male")
average_session_duration_by_age(female_data, "Female")
```

```python
# This cell aggregates the BMI average and outliers of both male and females

def bmi_statistics(dataframe, gender): 
    # Once again set counters to 0
    total_bmi = 0         # Variable to calculate average
    outliers = 0          # Stores the number of gym members with BMI < 18.5 or > 30
    underweight = 0       # Stores the number of underweight males (under 18.5 BMI)
    obese = 0             # Stores the number of overweight males (over 30 BMI)
    count = 0             

    for index, row in dataframe.iterrows():
        total_bmi += row['BMI']  
        if row['BMI'] < 18.5:    # If the gym member has a BMI below 18.5 then both the below variables value increases by 1
            underweight += 1
            outliers += 1        
        elif row['BMI'] > 30:    # If the male gym member has a BMI above 30 then both the below variables value increases by 1
            obese += 1
            outliers += 1        
        count += 1              

    # Calculate the average BMI, rounded to 2 decimal places and print all results
    average_bmi = round(total_bmi / count, 2)
    print(f"\nThe average BMI for {gender}s is:", average_bmi)
    print(f"Total number of BMI Outliers for {gender}s:", outliers)
    print(f"Number of Underweight {gender}s (BMI < 18.5):", underweight)
    print(f"Number of Obese {gender}s (BMI > 30): {obese} \n")

# Apply the function to male and female datasets
bmi_statistics(male_data, "Male")
bmi_statistics(female_data, "Female")


# After seeing initial results this section was created to get further insight
# Created a dictionary with the diffrent age groups as keys and associated values starting at 0
underweight_female_age_brackets = {
    "18-24": 0,
    "25-34": 0,
    "35-44": 0,
    "45-54": 0,
    "55+": 0
}

# For loop to go through the female dataframe and add values associated with relevant keys if the condition is met
for index, row in female_data.iterrows():
    if row['BMI'] < 18.5:  # Check if the BMI is underweight
        underweight_females = age_brackets(row['Age'])  # Determine the age bracket  by recalling predetermined function
        underweight_female_age_brackets[underweight_females] += 1

# Returns the summed up values with their associated age bracket
print("The number of underweight females in each age bracket are as follows:")
for underweight_females, underweight_count in underweight_female_age_brackets.items(): # .items() works in the same way as .iterrows but for dictionaries
    print(f"{underweight_females}: {underweight_count}")

```

```python
# This looks at the distribution of obese males accordng to BMI and males who are actually obese

# Create the empty dictionaries to store the values in 
bmi_obese_male_age_brackets = {
    "18-24": 0,
    "25-34": 0,
    "35-44": 0,
    "45-54": 0,
    "55+": 0
}

actually_obese_male_age_brackets = {
    "18-24": 0,
    "25-34": 0,
    "35-44": 0,
    "45-54": 0,
    "55+": 0
}

# Loop through the male dataframe
for index, row in male_data.iterrows():
    age_bracket = age_brackets(row['Age'])  # Determine the age bracket using the predefined 'age_brackets' function
    
    # If BMI is greater than 30, increase the counter for the age bracket
    if row['BMI'] > 30:
        bmi_obese_male_age_brackets[age_bracket] += 1
        
        # If BMI > 30 and Fat Percentage > 25, increase counter for the age bracket
        if row['Fat_Percentage'] > 25:
            actually_obese_male_age_brackets[age_bracket] += 1

# Prints results
print("The number of obese males (BMI > 30) in each age bracket are as follows:")
for age_bracket, count in bmi_obese_male_age_brackets.items():
    print(f"{age_bracket}: {count} \n")

print("\nThe number of obese males with BMI > 30 and Fat Percentage > 25 in each age bracket are as follows:")
for age_bracket, count in actually_obese_male_age_brackets.items():
    print(f"{age_bracket}: {count} \n")
```

This section of the notebook will further utilise functions and incorporate modules to decipher how to get the most out of workouts, specifically focusing on which workout type yields the best results.

As specific statistics on fitness goals such as muscle mass, flexibility, or endurance are unavailable, this analysis will use calories burned as the primary indicator of workout effectiveness. By comparing different workout types against the calories burned during sessions, we can determine which activities are the most efficient for energy expenditure. These insights can help participants optimise their workout routines.

```python
# This cell creates 3 new columns in the database. The most important being 'Calories_per_Minute'

# Create "Total_Calories_Burned_per_Week" column in dataframe
gym_members_data['Total_Workout_Time_per_Week (Mins)'] = (gym_members_data['Session_Duration (hours)'] * 
                                                          gym_members_data['Workout_Frequency (days/week)'] * 60)

# Create "Total_Calories_Burned_per_Week" column in dataframe
gym_members_data['Total_Calories_Burned_per_Week'] = (gym_members_data['Calories_Burned'] * gym_members_data['Workout_Frequency (days/week)'])

# Create "Calories_per_Minute" column in dataframe
gym_members_data['Calories_per_Minute'] = (gym_members_data['Total_Calories_Burned_per_Week'] / 
                                           gym_members_data['Total_Workout_Time_per_Week (Mins)'])

# Displays the firs 5 rows of the new columns
print(gym_members_data[['Total_Workout_Time_per_Week (Mins)', 'Total_Calories_Burned_per_Week', 'Calories_per_Minute']].head())
```

After running some initial tests in the notebook, it became clear that calories burned wasn't an accurate enough metric to see which workout type is best at burning calories as other metrics such as workout frequency and workout duration impacted results. As a result three new columns have been created "Total_Workout_Time_per_Week (Mins)", "Total_Calories_Burned_per_Week" but most importantly, "Calories_per_Minute".

The "Calories_per_Minute" metric offers a more standardised and fair comparison across workout types by accounting for differences in workout duration. This allows for a more accurate analysis of each workout's efficiency in calorie burning. By focusing on this refined metric, the aim is to identify the workout types that maximise calorie expenditure in the same amount of time.

```python

import numpy as np

# Makes a function that returns key numpy statistics for a specific column to be analysed
def numpy_statistics(dataframe, column):
    return {
        'Mean': np.mean(dataframe[column]),
        'Median': np.median(dataframe[column]),
        'Standard Deviation': np.std(dataframe[column]),
        'Min': np.min(dataframe[column]),
        'Max': np.max(dataframe[column])
    }

gym_members_data = pd.DataFrame(gym_members_data)

# Group data by Workout_Type
grouped_workout_types = gym_members_data.groupby('Workout_Type')

# Calculate statistics for each group and store them in a dictionary
results = {}
for workout_type, group in grouped_workout_types:
    # Apply numpy_statistics to the "Calories_per_Minute" column for the current group
    results[workout_type] = numpy_statistics(group, 'Calories_per_Minute')

print(f"The Following Represnts Calories Burned per Minute:\n")
# Print the results in a formatted manner
for workout_type, stats in results.items():
    print(f"Workout Type: {workout_type}")
    for stat, value in stats.items():
        print(f"{stat}: {round(value,2)}")
    print('\n')  # Add separator line between workout types
```

The analysis of calories burned per minute across different workout types reveals some interesting insights into workout intensity and effectiveness.

For Cardio workouts, the average calories burned per minute (mean) is 12.06, with a standard deviation of 1.4. This suggests that the calories burned during cardio sessions are relatively consistent, which is supported by the median value of 11.96 being very close. 

Surprisingly HIIT workouts have a slightly lower mean at 11.94 calories per minute and a standard deviation of 1.37. While the values are similar to cardio, the fact that HIIT sessions have a slightly lower average suggests that, accoridng to this data, it might burn slightly fewer calories per minute than cardio workouts.

Strength training shows the highest mean at 12.07 calories per minute, with a larger standard deviation of 1.57. This larger standard deviation suggests that there is greater variation in the calories burned during strength training sessions, with some participants burning considerably more or fewer calories than the average.

Yoga, despite being a lower-intensity workout, still burns a comparable amount of calories with a mean of 11.94 per minute. The standard deviation of 1.42 indiactes that there is some variation, but it is not as high as with strength training.

Despite all this the most ineteresting insight is the fact that all of the minimum and maximum values sit within 0.1 claories per minute of each other. This indicates that, irrespective of the workout type, the calorie expenditure across workouts remains fairly consistent.
